```python
import pandas as pd
filename='iris.csv'
iris=pd.read_csv(filename)
iris.head()
```

```python
iris_sorted = iris.sort_values(by='SepalWidth', ascending=False)
iris_sorted1 = iris_sorted.head(10)
```

```python
import matplotlib.pyplot as plt
iris_sorted1.plot(kind='bar', rot=0)
plt.title("top10")
plt.xlabel("data")
plt.show()
```

#연습4

```python
import pandas as pd
import matplotlib.pyplot as plt

titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
gender_survival = titanic.groupby('Sex')['Survived'].mean()
gender_survival
```

```python
parch_survival = titanic.groupby('Parch')['Survived'].mean()
parch_survival
```

```python
titanic1 = titanic.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
titanic2 = titanic1.groupby(['Pclass', 'Sex']).mean()
titanic2
```

```python
titanic3 = titanic.groupby(["Embarked","Survived"]).size().unstack()
titanic3
```

```python
import matplotlib.pyplot as plt
titanic3.plot(kind='bar', rot=0)
plt.xlabel('Embarked')
plt.ylabel('count')
plt.show()
```
