#4.판다스 데이터 선택


```python
#열 선택하기
import pandas as pd
df=pd.read_csv('/weather.csv')
df['temp']
df[['temp', 'date']]

#열 이름을 사용하여 열을 선택하려는 경우 loc()
df.loc[:,['date','temp']]

#열 인덱스를 사용하여 추출하려면 다음과 같이 iloc() 사용
df.iloc[:,[0,1]]

#특정 범위의 행 선택하기
df[0:3]

#특정 날짜에 해당하는 열 선택
df.index=df['date'] #기준=date 로 설정
df.loc['2010-08-01',['temp','mean_wind']]

#정수를 입력하면 해당 행 선택
df.iloc[3]

#특정 행과 열을 선택해서 슬라이싱
df.iloc[1:3,0:2]
#행,열 순서인건가..? 하지만 둘다 0번쨰 행/열을 포함함

```

```python
#불 인덱싱
w=df['temp']>=30
df[w]

#최고로 더웠던 날의 모든 정보를 출력
w=df['temp']==df['temp'].max()
df[w]

#기온이 30도 이상이고 최대풍속이 9 이상인 데이터
w=(df['temp']>=30) & (df['max_wind']>=9)
df[w]
#조건이 2개 이상이면 ()로 묶어줘야 함
```

#5. 판다스 결측데이터 처리
판다스는 누락된 데이터를 표시할 때 NaN으로 표기하며, 연산에는 포함되지 않는다.

```python
#결측데이터 확인

#최대풍속이 측정되지 않은 모든 데이터 출력
df[df['max_wind'].isna()]
```

```python
#결측데이터 개수 확인
df.isnull().sum()
```

#결측데이터 삭제하기 : dropna() 함수 사용

<매개변수>
-axis(축) : 0 또는 'index' 이면 누락된 값이 포함된 행 삭제, 1 또는 'columns' 이면 누락된 값이 포함된 열 삭제
-how : any 는 null 값이 있는 경우 해당 행 또는 열 삭제, all 은 모든 값이 누락된 경우 삭제
-inplace : True 로 설정하면 호출자 DataFrame 을 변경하는 부울 값 (?)

```python
#결측데이터가 있는 행 삭제 후 확인
df2=df.dropna()
df2.isnull().sum()
```

```python
#결측데이터 대체하기

df.fillna(df.mean(),inplace=True)
#inplace 를 True 로 설정해 원본 데이터 수정(?)

df.insull().sum()
```

#6. 판다스 데이터 가공

```python
dust=pd.read_excel('/dust1.xlsx')
dust.head()
```

```python
#컬럼/변수 삭제/생성하기

#삭제
#dust=dust.drop(['지역','망','측정소코드'], axis=1)
#dust.head()

#생성
dust['city']='서울'
dust.head()

#컬럼 이름 변경
dust.rename(columns={'측정소명':'name','측정일시':'date','주소':'addr'},inplace=True)
dust.columns
```

```python
#데이터형 확인
dust.dtypes
```

숫자 형식은 문자 형식으로 변환해야 함.

astype(str) 로 문자열 형식으로 변환 후 데이터 길이 날짜 형식 8자리에 적합하게 str.slice()를 사용하여 첫 8자리만 지정

```python
#숫자 int형을 문자열형으로 변환
dust['date']=dust['date'].astype(str)
dust['date']=dust['date'].str.slice(0,8)
dust.dtypes
```

```python
#문자열형을 날짜형으로 변환
dust['date']=pd.to_datetime(dust['date'])
dust.dtypes
```

```python
#날짜 형식 활용 : Series.dt. 형식

dust['year']=dust['date'].dt.year
dust['month']=dust['date'].dt.month
dust['day']=dust['date'].dt.day
dust.head()
```

**데이터 병합하기 **

- merge 함수
pd.merge(왼,오,how,on)
  - 아무 옵션도 적용하지 않으면 on : None 이고 두 데이터의 공통 열 이름(id)를 기준으로 inner(교집합) 조인
  - Outer 옵션을 적용하면 id를 기준으로 합치는데, 어느 한쪽에라도 데이터가 없는 경우 NaN값이 지정된다.
  - 왼/오른쪽에 입력한 데이터프레임 기준으로, 각각에 key 값에 해당하는 열을 지정한다.


```python
s1=pd.read_excel('/nation.xlsx')
s2=pd.read_excel('/code.xlsx')
s1, s2
```

```python
#공통 칼럼을 기준으로 병합
pd.merge(s1,s2,on='국적코드')
```

#7. 판다스 데이터 그룹핑

데이터 그룹핑은 데이터를 특정한 값을 기반으로 묶는 기능으로 groupby() 함수를 사용하여 통계량을 요약함
m  
- mean
- std(표준편차)
- var(분산)
- max
- min
- mode(최빈값)

```python
#한 열을 기준으로 그룹화
s1=pd.read_excel('/nation.xlsx')
s1.groupby('국적코드').sum()
```

```python
#여러 열을 기준으로 그룹화
s1.groupby(['국적코드', '성별']).sum()
```
